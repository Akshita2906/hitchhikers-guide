### Join the data

Each row of Court Commitments contains an inmate admission record. Instances where individual was incarcerated have a valid date under OFFENDER_ADMISSION/INTAKE_DATE and some sort of flag under INMATE_COMMITMENT_STATUS_FLAG. Instances not involving incarceration have OFFENDER_ADMISSION/INTAKE_DATE = 0001-01-01 and INMATE_COMMITMENT_STATUS_FLAG = NaN.

Each row of Inmate Profile contains individual-level information.

```python
cc_inmates = court_commit.merge(inmate, how='left', left_on="OFFENDER_NC_DOC_ID_NUMBER", right_on='INMATE_DOC_NUMBER')
```
